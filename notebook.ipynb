```python
import os

from pyspark.sql import SparkSession

db_url = "jdbc:mysql://localhost:3306/openfoodfacts"
db_properties = {
    "user": "root",
    "password": "root",
    "driver": "com.mysql.cj.jdbc.Driver",
}
mysql_driver_path = os.path.abspath("installation/mysql-connector-j-9.1.0.jar")
csv_filepath = os.path.abspath("data/openfoodfacts.csv")
```

```python
spark = (
    SparkSession.builder.appName("Int√©gration_donn√©es")
    .config("spark.jars", mysql_driver_path)
    .config("spark.driver.extraJavaOptions", "--add-opens java.base/java.nio=ALL-UNNAMED --add-exports java.base/sun.nio.ch=ALL-UNNAMED")
    .config("spark.executor.extraJavaOptions", "--add-opens java.base/java.nio=ALL-UNNAMED --add-exports java.base/sun.nio.ch=ALL-UNNAMED")
    .getOrCreate()
)

print("‚úÖ Spark initialis√© avec le driver MySQL :", mysql_driver_path)
print("üîó Driver charg√© :", spark.sparkContext.getConf().get("spark.jars"))
```

```python
df = spark.read.csv(
    csv_filepath,
    header=True,
    inferSchema=True,
    sep=" "
)

print(f"Nombre total de lignes charg√©es : {df.count()}")
df.printSchema()
df.describe().show(truncate=False)
df.limit(5).toPandas()
```
